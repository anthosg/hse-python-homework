```python
import datetime


class Account:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        self.transactions = []
        self.transactions.append([datetime.datetime.now(), self.balance])

    def increase(self, money):
        self.balance += money
        self.transactions.append([datetime.datetime.now(), money])
        return self.balance

    def decrease(self, money):
        if self.balance < money:
            print("Недостаточно средств. Баланс не изменился.")
            return self.balance
        self.balance -= money
        self.transactions.append([datetime.datetime.now(), money * -1])
        return self.balance

    def current_balance(self):
        return self.balance

    def current_transactions(self):
        print("Время операции              |  Сумма")
        print("----------------------------------------")
        for timed, money in self.transactions:
            print("{0}  |  {1}".format(timed, money))

    def current_name(self):
        return self.name


user = Account('Маугли', 1000)
print("Счет создан на имя {0} с балансом {1} рублей".format(user.current_name(), user.current_balance()))
print()

print("Добавим на счет 6000 рублей")
print("=" * 40)
print("Ваш текущий баланс:", user.increase(6000))
print()

print("Забираем со счета 5000 рублей")
print("=" * 40)
print("Ваш текущий баланс:", user.decrease(5000))
print()

print("Забираем со счета еще 3000 рублей")
print("=" * 40)
print("Ваш текущий баланс:", user.decrease(3000))
print()

print("История поступления или снятия денег")
print("=" * 40)
user.current_transactions()
```
